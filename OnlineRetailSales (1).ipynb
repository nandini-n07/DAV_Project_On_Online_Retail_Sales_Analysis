**Data Cleaning** 

Check the Null Values

```python
df.isnull().sum()
```

Description column is having null value so, we replace this null value with if any Stockcode having the NAN description we look for the other records having the most frequent Description value and replace them with that.

```python
freq=df[['StockCode','Description']].value_counts().reset_index()   
#freq
most_freq=freq.groupby('StockCode').head(1)
most_freq
```

We merge this most frequent description on the dataframe using stockcode on leftside

```python
most_freq.columns=['StockCode','Freq_Description','count']
df1=df.merge(most_freq,on='StockCode',how='left')
df1
```

Here we replace this frequent descrpition column with our actual description column and drop the NAN values in the Description column and also the frequent descrpition and count column

```python
df1['Description']=df1['Freq_Description']
df1.dropna(subset=['Description'],inplace=True)
df1.drop(columns=['Freq_Description','count'],inplace=True)
```

**Timely Analysis**

**1. How does Number of Transactions vary over month?**

```python
df2['Month']=df2['InvoiceDate'].dt.month
transactions_per_month=df2.groupby('Month')['InvoiceNo'].count()
transactions_per_month
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=transactions_per_month.index, y=transactions_per_month.values, marker='o', color='b')
plt.xlabel("Month")
plt.ylabel("Number of Transactions")
plt.title("Montly Transactions Over Time")
plt.grid(True)
```

* The highest number of transactions occurs in november month.
* In febraury least number of transaction is occured.
* From may to july a constant number of transaction is made.

**2. Identitfy montly sales trends in purchasing behaviour?**

```python
df2['UnitPrice']=df2['UnitPrice'].astype(int)
df2['Revenue']=df2['Quantity']*df2['UnitPrice']
monthly_sales = df2.groupby('Month')['Revenue'].sum().reset_index()
monthly_sales
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Revenue', marker='o', color='b')

plt.title('Monthly Sales Trends', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.7)
```

* We can notice may and june maintained same total revenue, this means a constant sales is made.
* In decemender a highest total revenue is made.

**3. How does revenue fluctuate throughout the day?**

```python
df2["Hour"] = df2["InvoiceDate"].dt.hour
hourly_revenue = df2.groupby("Hour")["Revenue"].sum().reset_index()
hourly_revenue
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x="Hour", y="Revenue", data=hourly_revenue, marker="o", color="b")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Revenue")
plt.title("Hourly Revenue Trend")
plt.xticks(range(0, 25))
```

* The line starts from 6 to 20, this means the store opens at morning 6'O clock and close at 8'O clock.
* The peak revenues is generated from 9:30 to 3:30 this means during this is peak working time.

**4. What are the peak sales hours and days of the week?** 

```python
df2["DayOfWeek"] = df2["InvoiceDate"].dt.strftime("%A")  # Day names
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Sunday"]
df2["DayOfWeek"] = pd.Categorical(df2["DayOfWeek"], categories=day_order, ordered=True)
heatmap_data = df2.groupby(["DayOfWeek", "Hour"])["Revenue"].sum().unstack()
plt.figure(figsize=(15, 6))
sns.heatmap(heatmap_data.loc[day_order], cmap="Blues", linewidths=0.5, annot=True, fmt=".0f")

plt.xlabel("Hour of the Day")
plt.ylabel("Day of the Week")
plt.title("Sales Heatmap: Peak Sales Hours & Days")
```

* The highest sales occur between 10 AM - 3 PM on most days.
* Tuesday (10 AM) shows the highest revenue, suggesting it might be a prime shopping day.
* Sales drop significantly after 4 PM, indicating fewer transactions in the evening.
* Tuesday, Wednesday, and Friday have the highest revenue, especially in mid-morning and early afternoon.
* Sunday shows significantly lower sales, meaning customers are less active on this day.

**Customer Analysis**

**5. Which customer generated most revenue?**

**6. What is customer Retention over time?**

```python
first_purchase=df2.groupby('CustomerID')['Month'].min()
df2["CohortMonth"] = df2["CustomerID"].map(first_purchase)
df2["CohortIndex"] = (df2["Month"] - df2["CohortMonth"])
cohort_counts = df2.groupby(["CohortMonth", "CohortIndex"])["CustomerID"].nunique().unstack()
#cohort_counts
cohort_sizes = cohort_counts.iloc[:, 0] 
#cohort_sizes
retention_matrix = cohort_counts.divide(cohort_sizes, axis=0) * 100 
retention_matrix
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(retention_matrix, annot=True, fmt=".1f", cmap="Blues", linewidths=0.5)
plt.title("Customer Retention Over Time")
plt.xlabel("Months Since First Purchase")
plt.ylabel("Cohort Start Month")
plt.yticks(rotation=0)
```

* High Drop-off in Early Months – Most customers stop purchasing after Month 0, with retention dropping to 30-40% in Month 1.
* Retention Spikes in Some Months – Temporary increases in retention (e.g., Month 4) may indicate successful promotions or seasonal trends.
* Some cohorts retain customers better – For example, Cohort 1 retains ~50% of customers by Month 10, much higher than later cohorts. 
  This suggests seasonal effects, strong marketing efforts.

**12. What is the average revenue per order across different countries?**

```python
order_revenue = df2.groupby(["InvoiceNo", "Country"])["Revenue"].sum().reset_index()
order_revenue
country_avg_revenue = order_revenue.groupby("Country")[["Revenue"]].mean().reset_index().sort_values(by="Revenue", ascending=False)
plt.figure(figsize=(18,6))
sns.barplot(x=country_avg_revenue["Country"], y=country_avg_revenue["Revenue"], color="b")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Average Revenue per Order ")
plt.title("Average Revenue per Order Across Different Countries")
plt.show()
```

* Singapore Has the Highest Average Revenue per Order – Orders from Singapore generate the most revenue per transaction, indicating high-value purchases.
* UK, Germany & France Have Lower Order Values – Despite being key markets in total revenue, their average revenue per order is relatively low, suggesting a high volume of smaller purchases.
* Saudi Arabia Has the Lowest Revenue per Order – This might indicate a preference for lower-cost items or smaller order sizes in that market.
